```python
df = pd.read_excel('.\sales.xlsx')
datalist = df['Profit'][(sales.Region == 'West')].tolist()
print (datalist)
```

```python
x = np.array(datalist)
StandDev = x.std()
print (StandDev)
```

```python
Sum = sum(datalist)
avg = Sum/len(datalist)
print(avg)
```

```python
ZscoreA = (5000-avg)/StandDev
print (ZscoreA)
```

```python
ZscoreB = (100-avg)/StandDev
print (ZscoreB)
```

```python
print(st.norm.cdf(ZscoreA) - st.norm.cdf(ZscoreB))
```
